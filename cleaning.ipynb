```python
#tokenize the name fields in fifa and plClean datasets
plClean['Name'] = plClean['Name'].str.split(" ")
fifa['long_name'] = fifa['long_name'].str.split(" ")

plClean['Name'] = plClean['Name'].apply(frozenset)
fifa['long_name'] = fifa['long_name'].apply(frozenset)
```

```python
llClean['name'] = llClean['name'].str.split(" ")
llClean['name'] = llClean['name'].apply(frozenset)
```

```python
#create a copy of the data for merging

fifa_copy = fifa

#iterate through PL data and match fifa player to current PL player
#update the name of the fifa player to reflect the PL data player name
for pl_row in plClean.iterrows():
    
    name = pl_row[1]['Name']
    club = pl_row[1]['Club']
    index_val = fifa_copy[(fifa_copy['long_name'].map(name.issubset)) & (fifa_copy['club'] == club)].index
    if len(index_val.tolist()) > 0:
        index_int = index_val.tolist()[0]
        
        fifa_copy.at[index_int, 'long_name'] = name
        
```

```python
#create a copy of the data for merging

fifa_copy = fifa

#iterate through PL data and match fifa player to current PL player
#update the name of the fifa player to reflect the PL data player name
for ll_row in llClean.iterrows():
    
    name = ll_row[1]['name']
    club = ll_row[1]['club']
    index_val = fifa_copy[(fifa_copy['long_name'].map(name.issubset)) & (fifa_copy['club'] == club)].index
    if len(index_val.tolist()) > 0:
        index_int = index_val.tolist()[0]
        
        fifa_copy.at[index_int, 'long_name'] = name
```

```python
#rename columns so that an outer join can be performed
fifa_copy = fifa_copy.rename(columns={'long_name' : 'name'} )
plClean = plClean.rename(columns={'Name' : 'name', 'Club':'club'})
```

```python
#ll merge on name and club
fifa_copy = fifa_copy.rename(columns={'long_name' : 'name'} )

resultll = pd.merge(llClean, fifa_copy, how="outer", on=['name', 'club'])
```

```python
#merge on name and club
result = pd.merge(plClean, fifa_copy, how="outer", on=['name', 'club'])
```
